## 1. Cinematica del movimiento

El auto es capaz de recorrer 10 metros en 2.5 segundos, por lo que deseamos conocer la velocidad que este alcanzara

```python
"Inputs" 
t = 2.5
d = 10 

"Calculo"
v = d/t

"Outputs"
print(f"Velocidad alcanzada {v:.3f} m/s")
```

## 2. Fuerzas de traccion

Se necesita determinar la fuerza necesaria para mover el vehiculo

```python
"Inputs"
mu = 0.1
m = 5
g = 9.81

"Calculo"
F=mu*m*g

"Outputs"
print(f"Fuerza necesaria{F:.3f} N")
```

## 3. Caracterizacion de la bateria

# 3.1 Potencia del motor

Se conoce que el motor consume 20 V y 5 A, por lo que se desea conocer la potencia de este motor electrico

```python
"Inputs"
V = 25
I = 10

"Calculo"
P = V*I

"Outputs"
print(f"Potencia del motor = {P:.3f} W")
```

# 3.2 Tiempo de vida de la bateria

Para conocer cuanto tiempo puede operar el vehiculo es necesario determinar el tiempo de vida antes de ser recargada

```python
"Inputs"
P = 250  #W
C = 500 #Wh

"Calculo"
t = C/P #horas

"Output"
print(f" el vehiculo puede circular por {t:.3f} horas")
```

## 4. Precio de recarga


Se desea conocer cuanto costaria recargar una vez la bateria

```python
"Inputs"
precio_kWh = 250 #CLP
Capacidad = 500 #Wh

"Calculos"
Capacidad = Capacidad/1000

"Output"
Costo = precio_kWh * Capacidad
```

## 5. Porcentaje de calentamiento

Se desea conocer que tan cerca de la temperatura maxima se encuentra el motor al estar trabajando

```python
"Inputs"
T_maxima_motor = 80
T_maxima_registrada = 71

"Calculo"
alcanzado = T_maxima_registrada/T_maxima_motor

"Output"
print(f"En condiciones de operacion se alcanzo un {alcanzado*100:.3f} % de la temperatura maxima")
```
